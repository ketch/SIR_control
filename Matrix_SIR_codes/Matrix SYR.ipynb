# SEIR

```python
beta = 0.3
gamma = 0.1
gamma_e = 0.2
b = np.array([0.,beta])
alpha = np.array([1., 0.])
V = np.array([[gamma_e, 0.],[-gamma_e,gamma]])
s0 = np.array(0.9)
y0 = np.array([0.01,0.05])
Vinv = np.linalg.inv(V)
R = b@Vinv@alpha
print(R)
bVi = b@Vinv
```

```python
s, y, t = SYR_forward(b,alpha,V,s0=s0,y0=y0,T=200)
Y = np.zeros_like(t)
for i in range(len(t)):
    Y[i] = bVi@y[:,i]/R
```

```python
plt.plot(t,s)
plt.plot(t,y[0,:])
plt.plot(t,y[1,:])
plt.plot(t,Y)
plt.legend(['S','E','I','Y'])
```
