Testing the characterization of SIR model via one first-order ODE

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def f1(t,x,x0,y0,beta,gamma):
    return -beta*x*(y0 + x0 - x) - gamma*x*np.log(x/x0)
```

```python
x0 = np.array([0.99])
y0 = 0.01
beta = 0.2
gamma = 0.1
T = 200

solution = solve_ivp(f1,[0,T],x0,t_eval=np.linspace(0,T,T),args=(x0,y0,beta,gamma))
```

```python
t = solution.t
x = solution.y[0,:]
plt.plot(t,x)
```

```python
y = y0 + (x0-x) + np.log(x/x0)*gamma/beta
plt.plot(t,y)
```
